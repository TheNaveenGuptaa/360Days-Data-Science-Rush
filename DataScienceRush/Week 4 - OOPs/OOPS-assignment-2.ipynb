```python
# Q1. What is Abstraction in OOps? Explain with an example.
"""
- An abstract class can be considered as a blueprint for other classes. 
- It allows you to create a set of methods that must be created within any child classes built 
  from the abstract class. and if a method is not declared within a class that is present in the abstract class , 
  then it will not create instances 
- Abstraction in python is defined as a process of handling complexity by hiding unnecessary 
  information from the user.
- simply , abstract class implements abstraction for hiding complexity 
example given below 
"""
from abc import ABC , abstractmethod
class car(ABC):     ### Creating a Abstract Base class for car details
    @abc.abstractmethod
    def mileage(self):
        pass
    @abc.abstractmethod
    def colour(self):
        pass
    
class baleno(car):
    def mileage(self):
        print('Mileage is 30 km/l')
    def colour(self):
        print('colour is grey')
        
class scorpio(car):
    def mileage(self):
        print('Mileage is 15 km/l')
    def colour(self):
        print('Color is black')
        
my_car = baleno()
his_car = scorpio()
my_car.mileage()
his_car.mileage()
my_car.colour()
his_car.colour()
```

```python
# Q5. Can we create an instance of an abstract class? Explain your answer.
"""
No , we can't create instance of abstract classes because it is not complete 
but it can be inherited to a child class of which instances can be created ...
"""
# example 
from abc import ABC, abstractmethod
class Animal(ABC):
 
    def move(self):
        pass
 
class Human(Animal):
 
    def move(self):
        print("I can walk and run")
 
class Snake(Animal):
 
    def move(self):
        print("I can crawl")

R = Human()
R.move()
 
K = Snake()
K.move()
```
