```python
## Q3. What errors are defined in the ArithmeticError class? Explain any two with an example.

logging.info('Problem 3 Started')
"""
Three Major errors are defined in the ArithmeticError class :-

1) FloatingPointError
2) OverflowError
3) ZeroDivisionError

"""
## OverflowError :- occurs when result is too large to be represented:
try: 
    import math
    print(math.exp(1000))
except OverflowError as e:
    logging.error(e)
## ZeroDivisionError :- occurs when we try to divide something with zero
try: 
    a = 5
    print(a/0)
except ZeroDivisionError as e:
    logging.error(e)
finally:
    logging.error('Problem 3 completed')
```

```python
## Q4. Why LookupError class is used? Explain with an example KeyError and IndexError.
logging.info('Problem 4 started')
"""
--> LookupError Exception is the Base class for errors raised when something can't be found.
--> It forms the base class for all exceptions that are raised when an index or a key is not 
found for a sequence or dictionary respectively.
"""

## KeyError : The KeyError is a type of LookupError exception and denotes that there was an issue retrieving the key you were looking for
try:
    d = {'key':'sudh',1:[1,2,3,4,5]}
    print(d['key2'])
except KeyError as e:
          logging.error(e)
## IndexError: It is a type of lookupError exception and is raised when an specified index of a sequence is not found 
try:
    l = [2,3,4,5]
    print(l[6])
except IndexError as e:
    logging.error(e)
finally:
    logging.info('Problem 4 Completed')
```
