```python
#### Problem 2
## using for loop
l = list(range(1,11))
s = 0
p = 1
for a in l :
    s+=a
    p*=a
print(s,p)
## using while loop 
o = 1
s1 = 0
p1 = 1
while o <= 10:
    s1 += l[(o-1)]
    p1 *= l[(o-1)]
    o+=1
print(s1,p1)
    

```

```python
### Problem 3 
u = eval(input('Welcome to Electricity Bill Calculator , Enter no. of Units Consumed : '))
cost = 0
for a in range(u+1):
    if a<=100 and a>0:
        cost+=4.5
    elif a<=200 and a>100:
        cost+=6
    elif a<=300 and a>200:
        cost+=10
    elif a>300:
        cost+=20
print(cost)
```

```python
### Problem 4 
nl = list(range(1,101))
l = []
for a in nl:
    if (a**3) % 4 == 0 or (a**3) % 5 == 0:
        l.append(a)
print(l)
```
